```python
import pandas as pd
```

```python
# Read in the data
file = 'Resources/Opioid_EMS_Calls.csv'

# Load the csv file into a pandas dataframe
opioid_df = pd.read_csv(file)
opioid_df.head()
```

```python
# Create a list of the columns
opioid_df.columns
```

```python
# Drop the Spec_Pop column
opioid_df = opioid_df.drop(columns=['Spec_Pop'])
opioid_df.head()
```

```python
# Create a new column for spec_pop based on the values in the ASU, Homeless, and Veteran columns
opioid_df['Spec_Pop'] = ''

# Loop through the rows and assign the appropriate value to the spec_pop column
spec_pop = []
for i, row in opioid_df.iterrows():
    if row['Patient_ASU'] == 'Yes' and row['Patient_Homeless'] != 'Yes' and row['Patient_Veteran'] != 'Yes':
        spec_pop.append('ASU Student')
    elif row['Patient_Homeless'] == 'Yes' and row['Patient_Veteran'] != 'Yes' and row['Patient_ASU'] != 'Yes':
        spec_pop.append('Homeless')
    elif row['Patient_Veteran'] == 'Yes' and row['Patient_Homeless'] != 'Yes' and row['Patient_ASU'] != 'Yes':
        spec_pop.append('Veteran')
    elif row['Patient_Homeless'] == 'Yes' and row['Patient_Veteran'] == 'Yes' and row['Patient_ASU'] != 'Yes':
        spec_pop.append('Veteran/Homeless')
    else :
        spec_pop.append('Other/Unknown')
```

```python
# Assign the list of values to the spec_pop column
opioid_df['Spec_Pop'] = spec_pop
opioid_df
```

```python
# Check the values in the spec_pop column
opioid_clean_df = opioid_df[['Patient_ASU', 'Patient_Homeless', 'Patient_Veteran', 'Spec_Pop']]
```

```python
# Export the dataframe to a .csv file in Resources
# Export without the headers for the sqlite import format
pd.DataFrame.to_csv(opioid_df, 'Resources/opioid_ems_calls_cleaned.csv', header=False, index=False)
```
